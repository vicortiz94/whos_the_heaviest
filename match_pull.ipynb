# Grabbing all match files

We previously got the match list of all the games i've played in the past year. Now, we want to grab the match data for each of those matches.

```python
from riotwatcher import LolWatcher, ApiError
import datetime
import json
import time
import os

api_key = ''

#lol_watcher is a wrapper for the Riot API
lol_watcher = LolWatcher(api_key)

#the servers are split into different regions. na1 signifies North America
my_region = 'na1'
```

This function will go through a matchlist file, grab the match_id, and then grab and save the match data. It will return the count to ensure that number of request we are performing still fall under the request limits set by Riot.

```python
def download_gamedata(filename, count):
    #I open the file and set the contents as data
    print('opening file')
    f_n = 'all_matchlist/%s'% filename
    with open(f_n) as json_file:
        data = json.load(json_file)
    
    #I'll go through each match, grab the match id, get the match data, and save it
    print('file opened, grabbing info')
    for matches in data['matches']:
        #grabbing match id
        match_id = matches['gameId']
        
        #there were multiple instances of 504 gateway errors, so I had it wait a while to try again to grab the data
        try:
            match_info = lol_watcher.match.by_id(my_region,match_id)
        except:
            print('trying again in one minute')
            time.sleep(60)
            print('trying again')
            match_info = lol_watcher.match.by_id(my_region,match_id)
        
        #I saved the data and placed it in the all_matches folder
        f_n2 = 'all_matches/%s_match.json'% match_id
        with open(f_n2, "w") as outfile: 
            json.dump(match_info, outfile)
    
        count += 1
        
        #count to check that we are under the request limit.
        if count % 15 == 0:
            print ('currently on ', count)
        #print(count)
        if count % 17 == 0:
            print('one second sleep')
            time.sleep(1)

        if count % 190 == 0:
            print('taking a two min break!')
            time.sleep(30)    
            print('90 sec')
            time.sleep(30)
            print('60 sec')
            time.sleep(30)
            print('30 more!')
            time.sleep(30)
            print('done!')
        
    return count
```

```python
count = 0

#this go through all the match list files we have
for filename in os.listdir('all_matchlist/'):
    print('starting', filename)
    c = download_gamedata(filename, count)
    count = c
    
    #as an extra precaution, had a break between files.
    print('taking a break! between files')
    time.sleep(30)    
    print('another 30 sec')
    time.sleep(30)
    print('done!')
```
